```python
b = electricity_path_df.set_index('Country Code')
elec_access = b.loc['WLD','1990':]

# Create a 2-Y axis graph
fig, ax1 = plt.subplots(figsize=(15, 8))

ax2 = ax1.twinx()
ax1.plot(x, life_exp, 'g-')
ax2.plot(x, elec_access, 'b-')

ax1.set_xlabel('Years')
ax1.set_ylabel('Life Expectancy (Years)', color='g')
ax2.set_ylabel('Access to electricity (% of population)', color='b')

# plt.xticks(rotation=90)
fig.autofmt_xdate(rotation = 90)

plt.savefig("Figures/elec_access.png")
plt.show()
```
